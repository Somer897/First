```python
def variable_1(a, b):
    diff =  len(a) - len(b)
    result = ''
    if diff == 0:
        for i in range(len(a)):
            result += chr(abs(ord(a[i]) - ord(b[i])))
    return result
```

```python
variable_1('awsd', 'asdw')
```

```python
def abc(b= 'Ghost')->str:
    return(f'Hello {b}')

abc('jhgcfjh2w  b')
```

```python
def abc(b:int = 'Ghost')->str:
    return(f'Hello {b}')

abc('l;vjjvn')
```

```python
def abc(age, b= 'Ghost')->str:
    return(f'Hello {b}, {age}')

abc(2)
```

```python
def arg(*args):
    return(f'Hello {args[0]}')    #каб было красива

arg('a')
```

```python
def arg(*args):
    data = ' '.join(args)
    return(f'Hello {data}')    #каб было красива

arg('a', 'b')
```

```python
def arg(*args):
    return(f'Hello {' '.join(args)}')    #каб было красива

arg('a', 'b', 'c', 'd')
```

```python
def kwargs(**kwargs):
    return(f'Hello {kwargs}')    #каб было красива

kwargs(n = 'a', k = 'b')
```
